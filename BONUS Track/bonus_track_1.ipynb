```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from mpl_finance import candlestick_ohlc
import pandas as pd
import matplotlib.dates as mpl_dates
```

```python
plt.style.use('ggplot')
df = pd.read_csv("alphabet_stock_data__1_.csv")
google = df.loc[:, ['Date', 'Open', 'High', 'Low', 'Close']]
filter1 = pd.DatetimeIndex(google['Date'])>='2020-04-01'
filter2 = pd.DatetimeIndex(google['Date'])<='2020-06-01'
google = google[filter1 & filter2]
google['Date'] = pd.to_datetime(google['Date'])
google['Date'] = google['Date'].apply(mpl_dates.date2num)
google = google.astype(float)

# creo Subplots
fig, ax = plt.subplots()
candlestick_ohlc(ax, google.values, width=0.6, colorup='green', colordown='red', alpha=0.8)


ax.set_xlabel('Date')
ax.set_ylabel('$ Price')
fig.suptitle('Daily Candlestick Chart of Google from 2020-04-01 to 2020-06-01')

date_format = mpl_dates.DateFormatter('%d-%m-%Y')
ax.xaxis.set_major_formatter(date_format)
fig.autofmt_xdate()

plt.show()
```
